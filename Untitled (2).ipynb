EXPLICACAO

Oi, estou trazendo dados de uma empresa com lojas em diferentes estados. Vamos analisar informações relevantes como vendas, faturamento, 
produtos e devoluções.

IMPORTACAO

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Definido uma variavel com que vai receber a lista de arquivos.
lst_arquivo = os.listdir('Vendas')
display(lst_arquivo)
```

```python
# Separando em duas variaveis de vendas e devolucoes, fazendo uma confirmacao idependente se a palavra-chave estiver Maiuscula ou Menuscula.
Vendas = []
Devolucoes = []
for arquivo in lst_arquivo:
    if ('Vendas' in arquivo) | ('Vendas' in arquivo.lower()) | ('Vendas' in arquivo.upper()) :
        Vendas.append(arquivo)
    elif ('Devolucoes' in arquivo) | ('Devolucoes' in arquivo.lower()) | ('Devolucoes' in arquivo.upper()):
        Devolucoes.append(arquivo)

display(Vendas)
print('________________________ ')
print('')
display(Devolucoes)
```

TRATAMENTO

```python
# dataFrame  de Vendas
total_Vendas = []
for arquivo in Vendas:     
    tabela = pd.read_csv(f'Vendas/{arquivo}')
    total_Vendas .append(tabela)
t_Vendas = pd.concat(total_Vendas , ignore_index=True)
t_Vendas.fillna('no')
```

```python
# dataFrame de Devolucoes
total_Devolucoes = []
for arquivo in Devolucoes:     
    tabela = pd.read_csv(f'Vendas/{arquivo}')
    total_Devolucoes .append(tabela)
t_Dev = pd.concat(total_Devolucoes , ignore_index=True)
t_Dev.fillna('no')
```

```python
#Por estado
def tab_est(tipo,onde):
    for arquivo in tipo:
        if onde in arquivo:
            tabela = pd.read_csv(f'Vendas/{arquivo}')
            display(tabela)

tab_est(Vendas, 'Rio de Janeiro')
```

CALCULO DO PRODUTO MAIS VENDIDO

Abaixo esta o codigo para ordenar e calcular o produto mais vendido.
No Caso o produto mais vendido foi o iPhone.

```python
tabela_produtos = t_Vendas.groupby('Produto').sum()
tabela_produtos = tabela_produtos[['Quantidade Vendida']].sort_values(by='Quantidade Vendida', ascending=False)
display(tabela_produtos)
```

```python
tabela_produtos.plot.bar()
```

Aqui criei um codigo para a pesquisa do produto mais vendido de cada Estado.

```python

def Venda_est(onde):
    for arquivo in Vendas:
        if onde in arquivo:
            tabela = pd.read_csv(f'Vendas/{arquivo}')
            tabela_produtos = tabela.groupby('Produto').sum()
            tabela_produtos = tabela_produtos[['Quantidade Vendida']].sort_values(by='Quantidade Vendida', ascending=False)
            display(tabela_produtos)
            tabela_produtos.plot.bar()

Venda_est('Rio de Janeiro')
```

PRODUTO QUE MAIS FATURO

Abaixo esta o codigo para o calculo do faturamento.

Adicionei uma Nova coluna para realizar o calculo entre a Quantidade Vendida e o Preco Unitario.

```python
t_Vendas['Faturamento'] = t_Vendas['Quantidade Vendida'] * t_Vendas['Preco Unitario']
```

Novamente e possivel ver que o iPhone foi o produto que mais Vendeu e Faturou.

Mas e possivel ver que a Camera mesmo vendendo menos que o Tablet e o SmartWatch, teve um faturamento maior.

```python
fat = t_Vendas.groupby('Produto').sum()
fat = fat[['Faturamento']].sort_values(by='Faturamento',ascending=False )
fat
```

```python
fat.plot.bar()
```
